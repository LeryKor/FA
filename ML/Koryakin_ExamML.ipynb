# Очистка данных

```python
data_train.isna().sum()
```

Удалим id колонки, а также колонки, которые содержат слишком много пустых значений

```python
data_train.drop(columns=["Loan ID", "Loan Status", "Customer ID", "Credit Score", "Annual Income", "Years in current job", "Months since last delinquent"], inplace = True)
```

Удалим оставшиеся строчки содержащие пустые значения

```python
data_train.dropna(axis=0, how="any", inplace = True)
```

```python
data_train.isna().sum()
```

Повторим тоже самое с тестовыми данными

```python
data_test.dropna(axis=0, how="all", inplace=True)
data_test.drop(columns=["Loan ID", "Customer ID", "Credit Score", "Annual Income", "Years in current job", "Months since last delinquent"], inplace = True)
data_test.dropna(axis=0, how="any", inplace = True)
```

```python
data_test.isna().sum()
```

Преобразуем строковые значения в числовые

```python
data_train.head(1)
```

```python
data_train["Term"] = pd.factorize(data_train["Term"])[0]
data_train["Home Ownership"] = pd.factorize(data_train["Home Ownership"])[0]
data_train["Purpose"] = pd.factorize(data_train["Purpose"])[0]
data_test["Term"] = pd.factorize(data_test["Term"])[0]
data_test["Home Ownership"] = pd.factorize(data_test["Home Ownership"])[0]
data_test["Purpose"] = pd.factorize(data_test["Purpose"])[0]
```

```python
data_train.head(5)
```

# Обучение моделей

```python
X_train, y_train = data_train.drop(columns=["Bankruptcies"]), data_train["Bankruptcies"]
X_test, y_test = data_test.drop(columns=["Bankruptcies"]), data_test["Bankruptcies"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

Обучаем модели (SVC слишком долго работала, поэтому я ее не взял).

```python
models= [KNeighborsClassifier(n_neighbors=5), KNeighborsClassifier(n_neighbors=10),
         MLPClassifier(hidden_layer_sizes=(150, ), activation='relu', solver='adam', alpha=0.0001, max_iter=500),
         DecisionTreeClassifier(), GaussianNB(), RandomForestClassifier(n_estimators=100, random_state=42),
         BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)]
for model in models:
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  print(model)
  print()
  scores = cross_val_score(model, X_train, y_train, cv=5)
  print(f"score = {model.score(X_test, y_test)}")
  print(f'cross-validation-score = {scores.mean()}')
  print()
```

Варнинги возникают, тк минорный класс содержит всего один элемент.


# Лучшие результаты (98,7%) по кросс валидации показал BaggingClassifier с параметрами estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42.
