```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("mushrooms.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
X, y = data.drop(columns=["class_p"]), data["class_p"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=3)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
model = SVC()
svc_params = {'kernel': ['linear', 'rbf', 'sigmoid'], 'C': np.logspace(0, 4, 5)}
svc_grid_search = GridSearchCV(estimator=model, param_grid=svc_params, cv=5)
svc_grid_result = svc_grid_search.fit(X_train, y_train)
print("Best parameters: ", svc_grid_search.best_params_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
model = KNeighborsClassifier()
knn_params = {'n_neighbors': [i for i in range(2, 6)], 'weights': ['uniform', 'distance']}
knn_grid_search = GridSearchCV(estimator=model, param_grid=knn_params, cv=5)
knn_grid_result = knn_grid_search.fit(X_train, y_train)
print("Best parameters: ", knn_grid_search.best_params_)
print("Best score: ", knn_grid_search.best_score_)
```
