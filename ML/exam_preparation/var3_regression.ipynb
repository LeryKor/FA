```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("PB_1996_2019_NH4.csv", sep=";")
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'],format= '%d.%m.%Y')
data['Date'] = data[['Date']].apply(lambda x: x[0].timestamp(), axis=1).astype(int)
```

```python
data
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
X, y = data.drop(columns=["NH4"]), data["NH4"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'fit_intercept': [True, False],
    'n_jobs' : [i for i in range(10)],
    'positive' : [True, False]
}

# Создание экземпляра модели Linear Regression
lr = LinearRegression()

# Подбор гиперпараметров с использованием GridSearchCV
grid_search = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Вывод лучших гиперпараметров и оценки производительности
print("Лучшие гиперпараметры:", grid_search.best_params_)
print("Лучшая оценка производительности:", grid_search.best_score_)

# Оценка производительности модели на валидационном наборе с лучшими гиперпараметрами
best_model = grid_search.best_estimator_
r2_score = best_model.score(X_test, y_test)
print("R2-оценка на валидационном наборе:", r2_score)
```
