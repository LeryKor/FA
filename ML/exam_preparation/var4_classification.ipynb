```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.isna().sum().sum()
```

```python
X, y = data.drop(columns=["infected"]), data["infected"]

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=3)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
```

```python
model = KNeighborsRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(f'accuracy = {accuracy_score(y_test, y_pred.round())}')
```

```python
from sklearn.preprocessing import PolynomialFeatures
def polyFunction(deg):
  poly = PolynomialFeatures(degree=deg)
  poly_features_train = poly.fit_transform(X_train)
  poly_features_test = poly.fit_transform(X_test)
  binary = KNeighborsRegressor()
  binary.fit(poly_features_train, y_train)
  y_pred = binary.predict(poly_features_test)
  print(f'accuracy = {accuracy_score(y_test, y_pred.round())}')
  return binary.score(poly_features_test, y_test)
```

```python
for j in [i for i in range(20)]:
  print(f'Степень полинома: {j}')
  polyFunction(j)
```

Какой полином лучше (тут 15)

```python
for col in X_train.columns:
  X_train2 = X_train.drop(columns=[col])
  X_test2 = X_test.drop(columns=[col])
  model = KNeighborsRegressor()
  model.fit(X_train2, y_train)
  y_pred = model.predict(X_test2)
  print(f'accuracy = {accuracy_score(y_test, y_pred.round())}')
```

Значения где-то хуже чем исходные, где-то лучше
